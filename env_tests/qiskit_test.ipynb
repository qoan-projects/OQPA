Hamiltonian:

$H=h(X_1+X_2)+J Z_1 Z_2$

Trotter formula:

$e^{-iH t}=(e^{-i h X_1 \Delta t}e^{-i h X_2 \Delta t}e^{-i J Z_1 Z_2 \Delta t})^n,$

where $\Delta t=t/n$.

```python
# System Specification
t=1
n_trotter=5
dt=t/n_trotter
n_qubits = 2
circ = QuantumCircuit(n_qubits)

h=1
J=1


# Test Circuit
for qubit in range(n_trotter):
    circ.rx(h*dt,0)
    circ.append(depol1,[0])
    
    circ.rx(h*dt,1)
    circ.append(depol1,[1])
    
    circ.cx(0,1)
    circ.append(depol12,[0,1])
    
    circ.rz(J*dt,1)
    circ.append(depol1,[1])
    
    circ.cx(0,1)
    circ.append(depol12,[0,1])
    


circ.measure_all()
print(circ)
```
